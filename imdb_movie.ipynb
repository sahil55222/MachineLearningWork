```python
clean_df = df.drop(['id','rating'], axis = 1)
```

```python
clean_df
```

```python
from sklearn.model_selection import train_test_split

rest, test = train_test_split(clean_df, test_size=0.1, random_state=4)
train, val = train_test_split(rest, test_size=0.1, random_state=4)
print('training data:', train.shape, 'validation data:', val.shape, 'test data:', test.shape)
```

```python
labels = ['action', 'adventure', 'animation', 'comedy', 'crime', 'drama',
       'fantasy', 'horror', 'mystery', 'romance', 'sci-fi', 'short',
       'thriller']

from tensorflow.keras.preprocessing.image import ImageDataGenerator

data_datagen = ImageDataGenerator(rescale=1./255, horizontal_flip=True)

train_data = data_datagen.flow_from_dataframe(
    train,
    x_col='image_rating',
    y_col=labels,  # name for the labels
    target_size=(224, 224),
    class_mode='raw',
    shuffle = True,
    batch_size=32
)

test_data = data_datagen.flow_from_dataframe(
    test,
    x_col='image_rating',
    y_col=labels,  # name for the labels
    target_size=(224, 224),
    class_mode='raw',
    shuffle = False,
    batch_size=32
)

val_data = data_datagen.flow_from_dataframe(
    val,
    x_col='image_rating',
    y_col=labels,  #name for the labels
    target_size=(224, 224),
    class_mode='raw',
    shuffle = False,
    batch_size=32
)
```

```python
# Visualizing the Train data generator
import matplotlib.pyplot as plt

for i in train_data.next():
    print(len(i))
    plt.imshow(i[3])
    break
```

```python
import tensorflow as tf
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import Dense

# Load the pre-trained EfficientNetB0 model
base_model = EfficientNetB0(include_top=False, weights='imagenet', input_shape=(224, 224, 3))

# Freeze the base model's weights
base_model.trainable = False

# Add custom classification layers on top of the base model
model = tf.keras.Sequential([
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(13, activation='sigmoid')  # Use sigmoid activation for multi-label classification
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python

# Use the one-hot encoded labels as target data
r = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10,
    steps_per_epoch=len(train_data),
    validation_steps=len(val_data),
    verbose=1
)
```

```python
# plot the loss
plt.plot(r.history['loss'], label='train loss')
plt.plot(r.history['val_loss'], label='val loss')
plt.legend()
plt.show()
plt.savefig('LossVal_loss')

# plot the accuracy
plt.plot(r.history['accuracy'], label='train acc')
plt.plot(r.history['val_accuracy'], label='val acc')
plt.legend()
plt.show()
plt.savefig('AccVal_acc_rating_EfficientNetB0')
```
